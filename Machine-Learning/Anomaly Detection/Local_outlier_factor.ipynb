```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=300, noise=0.1)
```

```python
X
```

```python
X_outlier = np.random.uniform(low=-3, high=3, size=(20,2))
X = np.vstack((X,X_outlier))
X
```

```python
plt.scatter(X[:, 0],X[:, 1])
```

```python
out = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = out.fit_predict(X)
y_pred
```

```python
X_inlier = X[y_pred == 1]
X_outlier = X[y_pred == -1]
```

```python
plt.title("Local Outlier Factor (LOF)")
plt.scatter(X_inlier[:, 0], X_inlier[:, 1], c='b', s=20, edgecolors='k', label='Normal Data Points')
plt.scatter(X_outlier[:, 0], X_outlier[:, 1], c='r', s=20, edgecolors='k', label='Outlier Data Points')
plt.legend()
plt.show()
```
