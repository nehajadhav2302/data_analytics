**1. What is a RESTful API?**

->

* A RESTful API (Representational State Transfer API) is a way for different software systems to communicate over the internet using standard HTTP methods.
* It uses standard HTTP methods like GET, POST, PUT, and DELETE.
* It follows a client-server architecture, where the client requests and the server responds.
* Data is typically exchanged in JSON format.

**2. Explain the concept of API specification.**

->

* An API specification is a detailed document or description that defines how an API behaves and how developers should interact with it.
* It describes available endpoints, HTTP methods (GET, POST, etc.), and data formats.
* It acts as a contract between frontend and backend teams.
* It outlines possible responses, including status codes and error messages.

**8. What is the difference between GET and POST HTTP methods?**

->

**GET**
* Used to retrieve data from the server.
* Sends data in the URL (query parameters).
* Data is visible in the URL (less secure).
* Has length limitations due to URL size.

**POST**
* Used to send data to the server to create or update resources.
* Sends data in the request body.
* Data is hidden from the URL (more secure).
*Can handle large amounts of data.

**19. What are HTTP status codes, and why are they important in a Flask API?**

* They indicate whether the request was successful, failed, or requires additional action.
* In a Flask API, status codes help communicate the outcome of API operations clearly to clients.

**21.  How would you secure a Flask API?**

->

* Securing a Flask API involves multiple layers to protect it from unauthorized access, data leaks, and common web vulnerabilities.

* Key approaches include:
1. Authentication & Authorization: Use token-based methods like JWT, OAuth2, or API keys to verify and control user access.
2. Use HTTPS: Always serve the API over HTTPS to encrypt data in transit.

**23. What is the role of Flask’s session object?**

->

* Flask’s session object is used to store data specific to a user’s session across multiple requests.
* It allows you to keep user-related information (such as login status, preferences, or temporary data) without the user needing to resend it every time.
* The session data is stored client-side in a cookie, but it is cryptographically signed and encrypted using the app’s secret key to ensure security and integrity.
* The session behaves like a dictionary (session['key'] = value) and persists data for the duration of the user’s session.
* Common use cases include managing user login states, shopping carts, and flash messages.

```html
<!DOCTYPE html>
<html>
<head>
    <title>Login Form</title>
</head>
<body>
    <h2>Login</h2>
    <form method="POST">
        <label>Username:</label>
        <input type="text" name="username"><br><br>

        <label>Password:</label>
        <input type="password" name="password"><br><br>

        <input type="submit" value="Login">
    </form>

    {% if error %}
        <p style="color:red; margin-top: 30px;">{{ error }}</p>
    {% endif %}

    {% if message %}
        <p style="color:green; margin-top: 30px;">{{ message }}</p>
    {% endif %}
</body>
</html>
```

```python
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route('/', methods=['GET', 'POST'])
def login():
    error = ''
    message = ''
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']

        if not username or not password:
            error = 'All fields are required.'
        elif len(password) < 6:
            error = 'Password must be at least 6 characters long.'
        else:
            message = f'Welcome, {username}!'
    
    return render_template('login.html', error=error, message=message)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
from flask import Flask, render_template, request, session, redirect, url_for
from flask_session import Session

app = Flask(__name__)

app.config["SECRET_KEY"] = "your_secret_key"
app.config["SESSION_PERMANENT"] = False
app.config["SESSION_TYPE"] = "filesystem"
Session(app)

@app.route('/', methods=['GET', 'POST'])
def login():
    error = ''
    message = ''
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']

        if not username or not password:
            error = 'All fields are required.'
        elif len(password) < 6:
            error = 'Password must be at least 6 characters long.'
        else:
            session['username'] = username
            message = f'Welcome, {username}!'
    
    return render_template('login.html', error=error, message=message)

@app.route('/welcome')
def welcome():
    print(session)
    if 'username' in session:
        return f"Welcome, {session['username']}!"
    return redirect(url_for('login'))

if __name__ == '__main__':
    app.run(debug=True)
```

**10. How do you handle errors in Flask (e.g., 404)?**

```html
<!DOCTYPE html>
<html>
<head>
    <title>Home - Flask App</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #f2f2f2;
            text-align: center;
            padding: 50px;
        }
        h1 {
            color: #2c3e50;
        }
    </style>
</head>
<body>
    <h1>Welcome to the Flask Project</h1>
    <p>This is the home page of your Flask application.</p>
</body>
</html>
```

```python
from flask import Flask, jsonify, render_template

app = Flask(__name__)

@app.route('/')
def home():
    return render_template("index.html")

@app.errorhandler(404)
def page_not_found(e):
    return jsonify(Error="Page not found."), 404

@app.errorhandler(403)
def forbidden_error(e):
    return jsonify(Error="Access forbidden."), 403

@app.errorhandler(500)
def internal_error(e):
    return jsonify(Error="Internal server error."), 500

if __name__ == '__main__':
    app.run(debug=True)
```
