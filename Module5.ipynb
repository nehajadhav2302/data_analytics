**1. What is the difference between a function and a method in Python?**

->

*  **Function**


1.   A function is a block of code that performs a specific task.
2.   It is defined using def and can exist independently (i.e., not bound to a class or object).


*   **Method**


1.   A method is a function that is associated with an object, usually defined inside a class.
2.   It always takes self as the first parameter (in instance methods), referring to the instance calling it.




````python
# Function Example
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
````

````python
# Method Example
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Alice"))
````

**6.  Explain the concept of generators in Python and how they are defined.**

->

* A generator function is a special type of function that returns an iterator object.
* Instead of using return to send back a single value, generator functions use yield to produce a series of results over time.
* This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.
* Syntax:
```python
def generator_function_name(parameters):
    # Your code here
    yield expression
    # Additional code can follow
```

````python
# Example
def count_up_to(max):
    num = 1
    while num <= max:
        yield num
        num += 1
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
````

**7. What are the advantages of using generators over regular functions?**

->
* **Memory Efficiency**: Generators only produce one item at a time and donâ€™t store the entire sequence in memory, making them much more memory-efficient than regular functions that return a full list or collection.

* **Lazy Evaluation**: Lazy evaluation means values are only computed when needed (i.e., on demand), which can save time when iterating over large datasets. The generator pauses when the value is returned and resumes when next() is called again.

* **Simplified Code**: Generators can replace complex iteration logic, avoiding the need to create temporary lists or collections to hold intermediate results.

* **Better Handling of Infinite Sequences**: Generators are perfect for representing infinite sequences since they don't store all values in memory at once. They compute each value on demand.

**8.  What is a lambda function in Python and when is it typically used?**

->

* A lambda function in Python is a small, anonymous function defined using the lambda keyword.
* It allows you to create functions without the need to formally name them, making it convenient for simple, short-term operations.
* Syntax:
```python
    lambda arguments: expression
```

* lambda: The keyword used to define the function.

* arguments: A comma-separated list of parameters (like in a regular function).

* expression: A single expression that the function evaluates and returns.

* **When to Use Lambda Functions**

1. When you need a small function for a single operation, and defining a full function with def seems overkill.
2. Used with map(), filter(), and reduce().
3. When you need a quick function to be passed to a method, lambda allows you to avoid defining a full function just for that purpose.

````python
# Example
add = lambda a, b: a + b
print(add(3, 5))
````

**9.  Explain the purpose and usage of the `map()` function in Python.**

->
* The map() function in Python is used to apply a given function to all items in an iterable (like a list or tuple) and return a new iterator with the results.
* The purpose of map() is to transform each element of an iterable using a function.
* It avoids the need for explicit loops, making code cleaner and more functional in style.
* Syntax:
```python
map(function, iterable)
```



````python
# Example
numbers = [1, 2, 3, 4]
squared = map(pow, numbers, [2, 2, 2, 2])
print(list(squared))
````

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

->

* **map()**: Applies a function to each item in an iterable and returns a new iterator with the results.
* **reduce()**: Reduces an iterable to a single value by repeatedly applying a function to the accumulated result and the next item.
* **filter()**: Filters items in an iterable using a condition function that returns True or False.

````python
# map() example
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))
````

````python
# reduce() example
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x * y, numbers)
print(result)
````

````python
# filter() example
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
````

**5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

````python
class fibonacci:
  def __init__(self,terms):
    self.terms=terms
    self.count=0
    self.a=0
    self.b=1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count < self.terms:
      if self.count == 0:
        self.count += 1
        return self.a
      self.count += 1
      self.a, self.b = self.b, self.a + self.b
      return self.a
    else:
      raise StopIteration
````

````python
fib = fibonacci(10)
for num in fib:
    print(num, end=" ")
````

**6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.**

````python
def power_of_two(num):
  for i in range(num+1):
    yield 2**i

res=power_of_two(4)
print(next(res))
print(next(res))
print(next(res))
````

**7.  Implement a generator function that reads a file line by line and yields each line as a string.**

````python
def read_file(file_path):
  with open(file_path,'r') as file:
    for line in file:
      yield line.strip()

for line in read_file('example.txt'):
    print(line)
````
