**1.  What is hypothesis testing in statistics?**

* Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data.
* It involves formulating two competing hypotheses and using sample data to determine which hypothesis is more likely to be true.
* Types of hypothesis testing:
  * Null Hypothesis
  * Alternate Hypothesis

**2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Null Hypothesis (H₀):**
- It is the default assumption that there is no effect, no difference, or no relationship in the population.
- It represents the status quo or a statement of no change.
- The goal of hypothesis testing is to assess whether there is sufficient evidence to reject the null hypothesis.
- Example:
  H₀: The new teaching method has no impact on student performance.

**Alternative Hypothesis (H₁ or Ha):**
- It is a statement that contradicts the null hypothesis.
- It proposes that there is an effect, a difference, or a relationship.
- It is what researchers aim to support through data.
- Example:
  H₁: The new teaching method improves student performance.

**3. What is the significance level in hypothesis testing, and why is it important?**

**Significance Level (α):**
- The significance level, denoted by α (alpha), is the threshold probability used to decide whether to reject the null hypothesis.
- It represents the maximum probability of making a Type I error, which occurs when we reject a true null hypothesis.
- Common values for α are 0.05, 0.01, or 0.10, with α = 0.05 being most widely used.

**Why is it Important?**
- It determines the critical value or decision boundary in hypothesis testing.
- It defines the confidence level:  
  For example, if α = 0.05, we are 95% confident in our decision.
- It helps control the rate of false positives, i.e., concluding there is an effect when there isn't one.
- The choice of α affects the power of the test and the interpretation of p-values.

**4. What does a P-value represent in hypothesis testing?**

- A p-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis (H₀) is true.
- It helps you determine the strength of evidence against the null hypothesis.
- A small p-value (typically ≤ 0.05) indicates strong evidence against H₀, so you may reject the null hypothesis.
- A large p-value (> 0.05) suggests weak evidence against H₀, so you fail to reject H₀.


**5.  How do you interpret the P-value in hypothesis testing?**

- The p-value helps you decide whether or not to reject the null hypothesis.
- If the p-value is less than or equal to the significance level (α), you reject the null hypothesis. This means the observed data is unlikely under the assumption that the null hypothesis is true, so the result is statistically significant.
- If the p-value is greater than the significance level, you fail to reject the null hypothesis. This means the observed data is consistent with the null hypothesis, and there is not enough evidence to support the alternative hypothesis.

**6. What are Type 1 and Type 2 errors in hypothesis testing?**

**Type 1 Error (False Positive):**
- Occurs when the null hypothesis (H₀) is rejected even though it is actually true.
- This means concluding that there is an effect or difference when there isn’t one.
- The probability of making a Type 1 error is denoted by alpha (α), the significance level.

**Type 2 Error (False Negative):**
- Occurs when the null hypothesis (H₀) is not rejected even though it is actually false.
- This means failing to detect a real effect or difference.
- The probability of making a Type 2 error is denoted by beta (β).

**7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

**One-Tailed Test:**
- Tests for the possibility of the relationship in only one direction.
- The alternative hypothesis (H₁) states that the parameter is either greater than or less than a certain value.
- The entire significance level (α) is placed in one tail of the distribution.
- Example:
 - H₀: μ = 100
 - H₁: μ > 100 (right-tailed) or H₁: μ < 100 (left-tailed)

**Two-Tailed Test:**
- Tests for the possibility of the relationship in both directions.
- The alternative hypothesis (H₁) states that the parameter is not equal to a certain value.
- The significance level (α) is split between both tails of the distribution.
- Example:
 - H₀: μ = 100
 - H₁: μ ≠ 100


**8. What is the Z-test, and when is it used in hypothesis testing?**

- The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means or between two sample means.

- It is based on the Z-distribution (standard normal distribution) and is used when the population standard deviation is known and the sample size is large (typically n ≥ 30).

**When is it used:**
- To compare a sample mean to a known population mean.
- To compare the means of two independent samples (two-sample Z-test).
- To test proportions (one or two proportions Z-test).


**9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

- The Z-score is a measure that tells how many standard deviations a data point is from the mean. In hypothesis testing, it is used to compare the observed value with the expected value under the null hypothesis.

- Formula for Z-score:
 **Z = (X̄ - μ) / (σ / √n)**

Where:
- X̄ is the sample mean
- μ is the population mean
- σ is the population standard deviation
- n is the sample size

What it represents:
- The Z-score shows how far the sample mean is from the population mean in terms of standard errors.
- It helps determine the probability of observing a value as extreme as or more extreme than the test statistic under the null hypothesis.
- It is used to calculate the p-value and make a decision to accept or reject the null hypothesis.

**10. What is the T-distribution, and when should it be used instead of the normal distribution?**

- The T-distribution is a probability distribution that is similar to the normal distribution but has heavier tails.
- It is used in hypothesis testing and confidence interval estimation when the sample size is small and the population standard deviation is unknown.
- The shape of the T-distribution depends on the degrees of freedom (df = n - 1)
- As the sample size increases, the T-distribution approaches the normal distribution

**When to use the T-distribution instead of the normal distribution:**
- When the sample size is small (typically n < 30)
- When the population standard deviation (σ) is unknown
- When the underlying population is approximately normally distributed.

**11.  What is the difference between a Z-test and a T-test?**

**Z-test:**
- Used when the population standard deviation (σ) is known
- Typically applied when the sample size is large (n ≥ 30)
- Based on the standard normal distribution (Z-distribution)
- Provides more precise results when population parameters are known

**T-test:**
- Used when the population standard deviation is unknown
- Typically applied when the sample size is small (n < 30)
- Based on the T-distribution, which has heavier tails
- Accounts for extra variability due to estimating the standard deviation from the sample

**12. What is the T-test, and how is it used in hypothesis testing?**

- The T-test is a statistical test used to determine whether there is a significant difference between the means of two groups, especially when the population standard deviation is unknown and the sample size is small.
- It is based on the T-distribution and helps assess whether observed differences could have occurred by random chance.

**Types of T-tests:**
1. One-sample T-test: Compares the mean of a single sample to a known value or population mean.
2. Two-sample T-test (independent): Compares the means of two independent groups.
3. Paired T-test: Compares means from the same group at different times or under two different conditions.

**How it is used in hypothesis testing:**
- State the null hypothesis (H₀) and alternative hypothesis (Ha)
- Calculate the sample mean, standard deviation, and standard error
- Compute the T-statistic using the formula:

 **T = (X̄ - μ) / (s / √n)**

  where:
  - X̄ is the sample mean
  - μ is the population mean under H₀
  - s is the sample standard deviation
  - n is the sample size

- Determine the degrees of freedom (df = n - 1)
- Compare the calculated T-statistic with the critical value from the T-distribution table, or find the p-value
- Make a decision: if p-value < α, reject H₀

**13. What is the relationship between Z-test and T-test in hypothesis testing?**

- Both tests compare observed data to what is expected under the null hypothesis.
- Both use a test statistic that measures how far the sample mean is from the population mean in standard error units.
- The main difference lies in whether the population standard deviation is known and the sample size.

**14. What is a confidence interval, and how is it used to interpret statistical results?**

- A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter with a certain level of confidence.

- It provides an estimate of the parameter (such as the mean or proportion) along with an indication of the uncertainty associated with the estimate.

- Common confidence levels are 90%, 95%, and 99%. A 95% confidence interval means that if we repeated the experiment many times, about 95% of the calculated intervals would contain the true population parameter.

**How it is used to interpret statistical results:**
- A confidence interval helps understand how precise an estimate is.
- If a confidence interval for the difference between two means includes zero, it suggests that there may be no significant difference.
- Narrow intervals indicate more precise estimates, while wider intervals suggest more variability or uncertainty.

Formula for confidence interval (for mean when σ is known):

**CI = X̄ ± Z * (σ / √n)**

Where:
- X̄ is the sample mean
- Z is the Z-score corresponding to the desired confidence level
- σ is the population standard deviation
- n is the sample size


**15. What is the margin of error, and how does it affect the confidence interval?**

- The margin of error is the maximum amount by which the sample estimate is expected to differ from the true population parameter. It reflects the uncertainty in the estimate due to sampling variability.

- It is a key component of the confidence interval and determines the range within which the true value is likely to lie.

- **Margin of Error (ME) = Z * (σ / √n)**

Where:
- Z is the Z-score for the desired confidence level (e.g., 1.96 for 95%)
- σ is the population standard deviation
- n is the sample size

**Effect on the confidence interval:**
- The confidence interval is calculated as:
  
  CI = Sample Estimate ± Margin of Error

- A larger margin of error results in a wider confidence interval, indicating more uncertainty.
- A smaller margin of error results in a narrower confidence interval, indicating more precision.

**16. How is Bayes' Theorem used in statistics, and what is its significance?**

- Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update the probability of a hypothesis based on new evidence or information.

- It provides a way to calculate the probability of an event, given prior knowledge of related conditions or events.

- Formula:

**P(A|B) = [P(B|A) * P(A)] / P(B)**

Where:
- P(A|B) is the probability of event A given that B has occurred (posterior probability)
- P(B|A) is the probability of event B given that A has occurred (likelihood)
- P(A) is the prior probability of event A
- P(B) is the total probability of event B

Significance:
- Allows for incorporation of prior knowledge or beliefs into the analysis
- Provides a dynamic framework for learning from data as new information becomes available
- Widely used in modern statistics, especially in Bayesian statistics and machine learning


**17. What is the Chi-square distribution, and when is it used?**

- The Chi-square distribution is a probability distribution that is used in statistics for testing relationships between categorical variables and for assessing goodness of fit.

- It is a right-skewed distribution that becomes more symmetric as the degrees of freedom increase.

- The Chi-square distribution is defined only for non-negative values and is determined by its degrees of freedom (df).

**When it is used:**
1. Goodness-of-fit test:
   - To determine whether an observed frequency distribution differs from a theoretical distribution.
   - Example: Checking if a die is fair by comparing the observed outcomes with expected frequencies.

2. Test of independence:
   - To assess whether two categorical variables are independent of each other.
   - Example: Testing if gender and preference for a product are related.

3. Test for variance:
   - To test hypotheses about population variance (in some advanced applications).

Formula for the Chi-square statistic:

**Chi-square (χ²) = Σ [(O - E)² / E]**

Where:
- O = Observed frequency
- E = Expected frequency


**18.  What is the Chi-square goodness of fit test, and how is it applied?**

- The Chi-square goodness of fit test is a statistical test used to determine whether the observed frequencies in a categorical dataset match the expected frequencies under a specific theoretical distribution.

- It helps to assess how well a sample data fits a distribution (such as uniform, binomial, or any hypothesized distribution).

**When to use:**
- When you have one categorical variable and want to check if the observed distribution fits a known or expected distribution.

**Steps to apply the test:**
1. Define the null hypothesis (H₀): The observed frequencies fit the expected distribution.
2. Define the alternative hypothesis (H₁): The observed frequencies do not fit the expected distribution.
3. Calculate the expected frequencies based on the theoretical distribution.
4. Compute the Chi-square statistic:

   **χ² = Σ [(O - E)² / E]**

   Where:
   - O = Observed frequency
   - E = Expected frequency

5. Determine the degrees of freedom (df = number of categories - 1).
6. Compare the calculated χ² value to the critical value from the Chi-square table, or use the p-value to make a decision.


**19. What is the F-distribution, and when is it used in hypothesis testing?**

- The F-distribution is a continuous probability distribution that arises frequently in the analysis of variance (ANOVA) and in testing whether two population variances are equal.

- It is defined as the ratio of two independent chi-square variables divided by their respective degrees of freedom.

- The F-distribution is positively skewed and depends on two degrees of freedom:
 - One for the numerator
 - One for the denominator

When it is used:
1. ANOVA (Analysis of Variance):
   - To compare the means of three or more groups.
   - Tests whether the group means are significantly different from each other.
   
2. Test for equality of variances:
   - Compares the variances of two populations to see if they are equal.

Formula for the F-statistic:

**F = (variance of group 1) / (variance of group 2)**

**20.  What is an ANOVA test, and what are its assumptions?**

- ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more independent groups to determine if there are significant differences between them.

- It tests the null hypothesis that all group means are equal against the alternative that at least one group mean is different.

**Types of ANOVA:**
- One-way ANOVA: Compares means based on one independent variable (factor).
- Two-way ANOVA: Compares means based on two independent variables.

**How it works:**
- ANOVA analyzes the variance within groups and the variance between groups.
- It calculates an F-statistic as the ratio of between-group variance to within-group variance.
- A large F-value suggests significant differences between group means.

**Assumptions of ANOVA:**
1. Independence:
   - The observations in each group must be independent of each other.

2. Normality:
   - The data in each group should be approximately normally distributed.

3. Homogeneity of variance (homoscedasticity):
   - All groups should have approximately equal variances.

If these assumptions are violated, the results of ANOVA may not be reliable. In such cases, alternative non-parametric tests (like the Kruskal-Wallis test) may be used.


**21. What are the different types of ANOVA tests?**

**1. One-way ANOVA:**
- Compares the means of three or more independent (unrelated) groups based on one factor (independent variable).
- Example: Comparing test scores of students from three different schools.

**2. Repeated Measures ANOVA:**
- Used when the same subjects are tested under different conditions or over time.
- Measures within-subject variation.
- Example: Measuring blood pressure of the same patients at different time intervals after medication.

**3. Factorial ANOVA**
- Factorial ANOVA is a type of analysis of variance used when there are two or more independent variables (factors), and you want to understand their individual (main) effects and combined (interaction) effects on a dependent variable.
- It is an extension of one-way ANOVA that allows the study of multiple factors simultaneously.

**22. What is the F-test, and how does it relate to hypothesis testing?**

- The F-test is a statistical test used to compare the variances of two or more groups and to assess whether the group means differ significantly in analysis of variance (ANOVA).

- It is based on the F-distribution, which is the ratio of two independent chi-square variables divided by their respective degrees of freedom.

**How it relates to hypothesis testing:**
- The F-test is commonly used in ANOVA to test the null hypothesis that all group means are equal.
- It can also be used to compare the variances of two populations (test of equal variances).
- The test statistic is calculated and compared to a critical value from the F-distribution table or used to compute a p-value.

**1. Write a Python program to generate a random variable and display its value.**

**2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).**

```python
low = 1
high = 7

dist = randint(low, high)

x = np.arange(low, high)

pmf_values = dist.pmf(x)

plt.stem(x, pmf_values, basefmt=" ")
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

**4.  Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.**

```python
n = 10
p = 0.5
size = 1000
samples = np.random.binomial(n, p, size)

plt.hist(samples, bins=range(n+2), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(range(n+1))
plt.show()
```

**5. Create a Poisson distribution and visualize it using Python.**

```python
lambda_val = 4
x = np.arange(0, 15)
pmf_val = poisson.pmf(x, mu = lambda_val)

plt.bar(x, pmf_val, color='skyblue', edgecolor='black')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability P(X = k)')
plt.title(f'Poisson Distribution (λ = {lambda_val})')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(x)
plt.show()
```

**6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.**

```python
low = 1
high = 7

dist = randint(low, high)

x = np.arange(low, high)

cdf_values = dist.cdf(x)

plt.step(x, cdf_values, where='post', color='green', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xticks(x)
plt.yticks(np.linspace(0, 1, high - low + 1))
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**7. Generate a continuous uniform distribution using NumPy and visualize it.**

```python
a = 2
b = 5
size = 10000

samples = np.random.uniform(low=a, high=b, size=size)

plt.hist(samples, bins=50, density=True, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Continuous Uniform Distribution U({a}, {b})')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**8.  Simulate data from a normal distribution and plot its histogram.**

```python
mu = 0
sigma = 1
size = 10000

samples = np.random.normal(loc=mu, scale=sigma, size=size)

plt.hist(samples, bins=50, density=True, color='lightcoral', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Normal Distribution (μ = {mu}, σ = {sigma})')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**9. Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std = np.std(data)

z_scores = (data - mean) / std

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=20, color='lightgreen', edgecolor='black', density=True)
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Density')
plt.grid(True)
```
